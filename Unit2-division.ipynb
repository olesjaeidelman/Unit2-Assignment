```python
print("This program checks if number divides by another number entered.")
prompt = "You will be asked for two numbers, press any key to continue or enter 'quit' to stop."
message = ""

active = True
while active:
  message = input(prompt)
  if message == 'quit':
    active = False
  else:
    prompt1 = "Please enter the number to check: "
    num = input(prompt1)
    prompt2 = "Please enter non-zero number to divide by: "
    check = input(prompt2)
    if num.isdigit() and check.isdigit() and check!=0 :
      if (int(num) % int(check)) == 0:
        message1 = 'Number: ' + num + ' divides by ' + check + '!'
      elif (int(num) % int(check)) != 0:
        message1 = 'Number: ' + num + ' and ' + check + ' are not divisible!'
    else:
      message1 = 'You entered: ' + num + ' and ' + check+ ". This is not a valid entry! Please try again."
    print(message1)
```
