```python
print("This programs plays Rock-Paper-Scissors")
prompt = "Two players will be asked to enter their selection, press any key to continue or enter 'quit' to stop."
message = ""

active = True
while active:
  message = input(prompt)
  if message.lower() == 'quit':
    active = False
  else:
    prompt1 = "First player, please enter rock, paper or scissors: "
    player1 = input(prompt1).lower()
    prompt2 = "Second player, please enter rock, paper or scissors: "
    player2 = input(prompt2).lower()
    if player1 in ('rock','paper','scissors') and player2 in ('rock','paper','scissors'):
      if player1 == 'rock' and player2 =='scissors':
        message1 = 'Player 1, you win!'
      elif player1 == 'scissors' and player2 =='paper':
        message1 = 'Player 1, you win!'
      elif player1 == 'paper' and player2 =='rock':
        message1 = 'Player 1, you win!'
      elif player1 == 'rock' and player2 =='rock':
        message1 = 'You picked the same, try again!'
      elif player1 == 'paper' and player2 =='paper':
        message1 = 'You picked the same, try again!'
      elif player1 == 'scissors' and player2 =='scissors':
        message1 = 'You picked the same, try again!'
      else:
        message1 = 'Player 2, you win!'
    else:
      message1 = 'You entered: ' + player1 + ' and ' + player2+ ". This is not a valid entry! Please try again."
    print(message1)
```
